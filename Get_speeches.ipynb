```python
def get_fields(ps):
    """
    parses AA speech fields
    
    Args
        ps: list of paragraph elements on page
        
    Returns:
        dict with speech information
    """
    
    Fields = ['Year','Category','Film Title','Winner']
    FieldDict = {}  # for storing information
    
    for p in ps:
        pp = str(p)
        
        # get field
        cat_re = re.search("(?<=>)(.*?)(?=:)",pp)
        if cat_re: cat = m.group()
            
        # get field content
        if cat in Fields:
            cat_value = re.search("(?<=<strong>)(.*?)(?=</strong>)",pp)
            if cat_value: FieldDict[cat] = cat_value.group()

        # get speech
        if re.compile(r'class="MInormal"').search(pp):
            # clean up speech string
            cleanstr = pp.replace('<br>', '').replace('\r','').replace('\n','').replace('</br>','').replace('<br/>','')
            
            # get speech
            speech = re.search("(?<=:)(.*?)(?=</p>)",cleanstr)
            if speech: FieldDict['Speech'] = speech.group()
    return FieldDict
```

```python
def processor(category,driver):
    """
    Runs through every category in chunks and crawls AA speeches
    
    Args
        category: name of award category
        driver: selenium driver
        
    Returns:
        list of category speeches
    """
    
    elem_length = len(driver.find_elements_by_link_text(category))
    content_list = []  # for storing records
    
    for i in range(elem_length):
        # fresh page elements
        elems = driver.find_elements_by_link_text(category)
        
        # sometimes it misses a link
        try:
            elems[i].click()
            
            # grab data
            url = urllib2.urlopen(driver.current_url)
            soup = BeautifulSoup(url, 'html.parser')
            ps = soup.find_all('p')
            content_list.append(get_fields(ps))
            
        except:
            print("category didnt work: %s"%category)
            continue
            
        time.sleep(0.1)
        
        # go back
        driver.back()
        time.sleep(0.1)
    return content_list
```

# put it all together

```python
All_df = [] # for storing df lists

for group in category_dict.keys():
    # setup
    kk = category_dict[group]
    url = kk['url']
    categories = kk['categories']
    
    # go to webpage
    driver = webdriver.Safari()
    driver.get(url)
    
    # get content
    startT = time.time()
    for category in categories:
        All_df.append(processor(category,driver))
        
    driver.close()
    print(startT - time.time())

dfs = []
for i in All_df:
    dfs.append(pd.DataFrame(i))
bigdf = pd.concat(dfs)
bigdf.to_csv('All_Oscars_Speeches.csv')
```
